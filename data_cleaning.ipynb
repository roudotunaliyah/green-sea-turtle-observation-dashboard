```python
import pandas as pd

df = pd.read_excel("Chelonia mydas.xlsx")
df.head()
```

```python
df = df.dropna(subset=["decimalLatitude", "decimalLongitude"])
df.head()
```

```python
df = df[(df["decimalLatitude"] != 0) & (df["decimalLongitude"] != 0)]
df.head()
```

```python
df = df.drop_duplicates(subset=["gbifID"])
df.head()
```

```python
df = df[df["occurrenceStatus"] == "PRESENT"]
df.head()
```

```python
df = df.dropna(subset=["year"])
df.head()
```

```python
df.to_csv("Chelonia_mydas_cleaned.csv", index=False)
```

```python
import pandas as pd

df_clean = pd.read_csv("Chelonia_mydas_cleaned.csv", low_memory=False)
df_clean.head()
```

```python
df_clean.columns
```

```python
keep_cols = [
    "gbifID",
    "decimalLatitude",
    "decimalLongitude",
    "countryCode",
    "stateProvince",
    "locality",
    "year",
    "month",
    "day",
    "eventDate",
    "scientificName",
    "basisOfRecord",
    "coordinateUncertaintyInMeters",
    "issues",
    "license",
    "institutionCode"
]

# Pastikan hanya kolom yang benar-benar ada
existing_cols = [col for col in keep_cols if col in df_clean.columns]

df_tidy = df_clean[existing_cols].copy()
df_tidy.head()
```

```python
df_tidy.to_csv("Chelonia_mydas_for_tableau.csv", index=False)
```
